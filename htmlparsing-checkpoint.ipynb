```python
from HTMLParser import HTMLParser as hp

class Parser(hp, object):
    def __init__(self):
        self.skip = True
        super(Parser, self).__init__()
        self.rowdata = []
        
    def handle_starttag(self, tag, attrs):
        if len(self.rowdata) >= 10:
            self.skip = True
        elif tag == 'tr':
            self.skip = False
            self.rowdata.append([])

    def handle_data(self, data):
        if not self.skip:
            self.rowdata[-1].append(data)

        
fi = open('/home/jamougha/Downloads/data.html', 'r')
htm = ''.join(line for line in fi.readlines())
#print htm
parser = Parser()
parser.feed(htm)
#print mydata[:1029]
print parser.rowdata[0][:100]
```

```python
def mungampersands(l):
    i, lenl, output = 1, len(l), [l[0]]

    while lenl - i > 2:
        if l[i] == '&':
            l[i - 1] = l[i - 1] + l[i] + l[i + 1]
            i += 2
        else:
            output.append(l[i])
            i += 1
    
    return output

rowonefix = mungampersands(parser.rowdata[1])
```

```python
from collections import defaultdict

def tokenize(ss):
    tokens = defaultdict(int)
    for s in ss:
        for token in s.split(' '):
            tokens[token] += 1
            
    tokpairs = ((v, k) for k, v in tokens.items())
    return sorted(tokpairs, reverse=True)

tokens = tokenize(rowonefix)
print len(tokens)
print tokens[:40]
    
```
