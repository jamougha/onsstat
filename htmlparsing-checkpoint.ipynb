```python
from six.moves import map, filter, range, zip
from __future__ import print_function, division, absolute_import, unicode_literals
import codecs
import re
```

```python
from HTMLParser import HTMLParser as hp

_mangedamp = '(╯°□°)╯︵ ┻━┻'

class Parser(hp, object):
    """Attempts to parse the broken HTML (neé XLS) files from
    the ONS website. Only succeeds for the header info but
    we can grab the rest from the .csv files. 
    """
    def __init__(self):
        self.skip = True
        super(Parser, self).__init__()
        self.rowdata = []
        
    def handle_starttag(self, tag, attrs):
        if tag == 'tr':
            self.skip = False
            self.rowdata.append([])

    def handle_data(self, data):
        if not self.skip:
            self.rowdata[-1].append(data)

def mungamps(s):
    """fix broken inclusion of ampersands in the html files
    Also HTMLParser is barfing on &amp; ...
    """
    mangled = '┬─┬﻿ ノ( ゜-゜ノ)'
    while s.find(mangled) != -1:
        mangled += mangled
        
    ampre = re.compile(r'&(nbsp;|lt;|gt;|quot;|cent;|pound;|yen;|euro;|copy;|lsquo;|rsquo;|ldquo;|rdquo;|times;|divide;|reg;|trade;)')
    def mangleamps(match):
        return mangled + match.groups()[0]
    
    smang = ampre.sub(mangleamps, s)
    return smang.replace('&', _mangedamp).replace(mangled, '&')

def mungfile(f = 'data.html'):
    """turn an ons xls (actually HTML) file into a table
    of values"""
    parser = Parser()        
    with codecs.open(f, 'r', 'Latin1') as fi:
        htm = ''.join(line for line in fi.readlines())
        parser.feed(mungamps(htm))
        
    data = [[elem for elem in row if elem.strip()] 
              for row in parser.rowdata]
    return data
```

```python
def iscdid(cdid):
    "Returns true if a string is a valid cdid. May or may not return false if it's not."
    return (len(cdid) == 4 and 
           any(x.isalpha() for x in cdid) and
           all(x.isdigit() or x.isalpha() and x.isupper() for x in cdid))
        
```

```python
def allyears(ps):
    for p in ps or []:
        if (len(p) != 4 or
           any(not c.isdigit() for c in p)):
            return None
    return ps or None
        
def interpret_table(data):
    """Take a table parsed from an ONS HTML file and extract a list 
    of the names of the columns, the corresponding CDIDs, and the 
    base and index periods if any.
    We don't extract the data itself because the HTML is too broken to 
    parse easily; we use the corresponding .csv for that.
    """
            
    names1 = max(data, key=lambda x: sum(map(len, x)))
    names = [x.replace(_mangedamp, '&') for x in names1]
    headdata = [x for x in data if len(x) == len(names)]

    def maybe_cdid_row(row):
        return (all(map(iscdid, row)) and 
               any(any(c.isdigit() for c in elem) for elem in row))
    cdids = [x for x in headdata if maybe_cdid_row(x)]
    
    if len(cdids) == 1:
        cdids, = cdids
    else:
        raise ValueError("Couldn't identify CDIDs: " + 
                        ("too many options." if len(cdids) == 0 else
                         "no valid rows"))
    
    base_period, index_period = None, None
    for now, nxt in zip(headdata, headdata[1:]):
        if now[0].lower() == 'base_period':
            base_period = nxt
        elif now[0].lower() == 'index_period':
            index_period = nxt  

    base_period, index_period = allyears(base_period), allyears(index_period)

    return names, cdids, base_period, index_period
```
